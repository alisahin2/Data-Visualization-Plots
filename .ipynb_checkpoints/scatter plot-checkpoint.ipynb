```python
#
import seaborn as sns
import matplotlib.pyplot as plt
ds = sns.load_dataset("tips")
```

```python
ds.head()
```

```python
sns.scatterplot(x = "tip", y = "total_bill", data = ds)
```

```python
sns.scatterplot(x = "tip", y = "total_bill", data = ds, hue = "sex")
```

```python
sns.scatterplot(
    x = "total_bill",
    y = "tip",
    hue = "day",
    style = "time",
    data = ds
)
```
